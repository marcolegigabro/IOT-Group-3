```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Rivar\Downloads\antoineangry.csv")
```

```python

df.head()
```

```python
df['_time'] = pd.to_datetime(df['_time'])
df = df.sort_values('_time')

plt.figure(figsize=(12, 6))
plt.plot(df['_time'], df['temperature'], label='Temperature', color='orange')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title("Temperature over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
selected_day = '2025-03-16'  # Change to your target date
mask = df['_time'].dt.date == pd.to_datetime(selected_day).date()
df_day = df[mask]

plt.figure(figsize=(12, 6))
plt.plot(df_day['_time'], df_day['temperature'], label='Temperature', color='orange')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title(f"Temperature on {selected_day}")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_hourly = df.set_index('_time').resample('1H').mean()
plt.figure(figsize=(12, 6))
plt.plot(df_hourly.index, df_hourly['temperature'], label='Temperature (1H avg)', color='orange')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title(f"Hourly Average Temperature")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
start_date = '2025-04-19'
end_date = '2025-04-26'

mask = (df['_time'] >= start_date) & (df['_time'] < end_date)
df_week = df[mask]

df_hourly = df_week.set_index('_time').resample('1H').mean()

plt.figure(figsize=(14, 6))
plt.plot(df_hourly.index, df_hourly['temperature'], label='Temperature (1H avg)', color='orange')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title(f"Hourly Average Temperature from {start_date} to {end_date}")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
